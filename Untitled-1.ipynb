```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from catboost import CatBoostClassifier
import datetime


# ================================================
# DATA PROCESSING MODULE
# ================================================

class DataProcessor:
    """Handles data loading, preprocessing and feature engineering"""
    
    def __init__(self):
        self.catboost_model = None
        
    def load_catboost_model(self, model_path):
        """Load the CatBoost model for predicting task durations"""
        self.catboost_model = CatBoostClassifier()
        self.catboost_model.load_model(model_path)
        print(f"CatBoost model loaded from {model_path}")
        
    def process_data(self, df_projects, df_pic_tasks, df_skills):
        """Process input dataframes to prepare for scheduling"""
        # Merge dataframes
        df_predict_takt = pd.merge(df_projects, df_pic_tasks, on='PRODUCT-ASSY TYPE Task', how='inner')
        df_predict_takt = df_predict_takt.merge(df_skills, on='PIC', how='inner')
        
        # Extract job structure
        jobs = df_projects.groupby('PN', sort=False)['PN Task'].apply(list).to_dict()
        
        # Get unique workers
        workers = list(df_predict_takt['PIC'].unique())
        
        # Create a mapping of tasks that shouldn't be assigned to the same worker
        not_same_pic = self._extract_not_same_pic_constraints(df_projects)
        
        return df_predict_takt, jobs, workers, not_same_pic
    
    def _extract_not_same_pic_constraints(self, df_projects):
        """Extract constraints where tasks shouldn't be assigned to the same worker"""
        not_same_pic = {}
        
        # Look for drawing tasks and final checking tasks that should have different workers
        for pn in df_projects['PN'].unique():
            drawing_tasks = df_projects[(df_projects['PN'] == pn) & 
                                       (df_projects['PN Task'].str.contains('DRAWING'))]['PN Task'].tolist()
            
            checking_tasks = df_projects[(df_projects['PN'] == pn) & 
                                        (df_projects['PN Task'].str.contains('FINAL CHECKING'))]['PN Task'].tolist()
            
            # Create constraints that drawing and checking should have different PICs
            for check_task in checking_tasks:
                for draw_task in drawing_tasks:
                    not_same_pic[check_task] = draw_task
        
        return not_same_pic
    
    def predict_processing_times(self, df_predict_takt):
        """Use CatBoost model to predict task processing times"""
        if self.catboost_model is None:
            raise ValueError("CatBoost model not loaded. Call load_catboost_model first.")
        
        # Prepare features for prediction
        cat_cols = ['Task', 'ASSY TYPE', 'PARTS TYPE', 'ORDER TYPE', 
                    'SITE_TYPE', 'PN_PREFIX', 'Designation']
        
        df_ml_dataset = df_predict_takt[[
            'Task', 'ASSY TYPE', 'PARTS TYPE',
            'ORDER TYPE', 'SITE_TYPE', 'PN_PREFIX',
            'JAGUAR', 'WCSP', 'STRIP', 'BI', 'TI', 'NON TI',
            'Designation',
            'CadExp_Enplas', 'Enplas_Years', 'CADWork_NonEnplas',
            'NonCadWork', 'Total',
            'NUM_SITE', 'NUM_STRIP'
        ]].copy()
        
        # Convert categorical columns
        for col in cat_cols:
            df_ml_dataset[col] = df_ml_dataset[col].astype('category')
        
        # Validate data for prediction
        if df_ml_dataset.empty:
            raise ValueError("Input DataFrame for prediction is empty.")

        missing_cols = set(self.catboost_model.feature_names_) - set(df_ml_dataset.columns)
        if missing_cols:
            raise ValueError(f"Missing features required by the model: {missing_cols}")
        
        # Predict processing times
        predicted_takt = self.catboost_model.predict(df_ml_dataset, prediction_type="RawFormulaVal")
        df_predict_takt['Predicted Takt'] = predicted_takt
        
        # Convert to the format needed by the scheduler
        processing_times = {}
        for task, group in df_predict_takt.groupby('PN Task'):
            processing_times[task] = dict(zip(group['PIC'], group['Predicted Takt']))
        
        return processing_times
    
    def calculate_delivery_based_deadlines(self, jobs, processing_times, df_projects):
        """Calculate deadlines based on delivery dates"""
        deadlines = {}
        today = datetime.datetime.now()
        
        for job, tasks in jobs.items():
            job_data = df_projects[df_projects['PN'] == job]
            
            if not job_data.empty and 'DeliveryDate' in job_data.columns and not pd.isna(job_data['DeliveryDate'].iloc[0]):
                # Use delivery date if available
                delivery_date = job_data['DeliveryDate'].iloc[0]
                days_until_delivery = (delivery_date - today).days
                # Convert days to time units
                time_units = days_until_delivery * 8  # Assuming 8 time units per day
                deadlines[job] = max(time_units, 20)  # Ensure minimum deadline
            else:
                # Calculate from task durations as fallback
                base_time = 0
                for task in tasks:
                    if task in processing_times and processing_times[task]:
                        min_time = min(processing_times[task].values())
                        base_time += min_time
                
                deadlines[job] = max(base_time * 1.5, 20)
                
        return deadlines
    
    def process_ongoing_tasks(self, ongoing_tasks, jobs, deadlines, processing_times):
        """Process ongoing tasks and update jobs, deadlines, and processing times"""
        updated_jobs = jobs.copy()
        updated_deadlines = deadlines.copy()
        updated_processing_times = processing_times.copy()
        
        for job, task, pic, takt in ongoing_tasks:
            if job in updated_jobs:
                # Add to existing job - make it the first task
                updated_jobs[job] = [task] + updated_jobs[job]
            else:
                # Create new job with this task
                updated_jobs[job] = [task]
                
                # Set deadline based on product type and estimated duration
                updated_deadlines[job] = takt * 5  # Simple heuristic: 5x task duration
            
            # Add processing time for this task/worker combination
            if task not in updated_processing_times:
                updated_processing_times[task] = {}
            updated_processing_times[task][pic] = takt
            
        return updated_jobs, updated_deadlines, updated_processing_times
```

```python
class FJSP_Env:
    """Flexible Job Shop Scheduling Environment"""
    
    def __init__(self, jobs, workers, processing_times, deadlines, 
                 switch_penalty=2, ongoing_tasks=None,
                 not_same_pic=None):
        self.jobs = jobs
        self.workers = workers
        self.processing_times = processing_times
        self.deadlines = deadlines
        self.switch_penalty = switch_penalty
        self.ongoing_tasks = ongoing_tasks
        self.not_same_pic = not_same_pic
        self.reset()

    def create_copy(self):
        """Create a deep copy of the environment for simulation"""
        env_copy = FJSP_Env(
            jobs=self.jobs,
            workers=self.workers,
            processing_times=self.processing_times,
            deadlines=self.deadlines,
            switch_penalty=self.switch_penalty,
            ongoing_tasks=self.ongoing_tasks,
            not_same_pic=self.not_same_pic
        )
        env_copy.schedule = list(self.schedule)
        env_copy.completed_jobs = set(self.completed_jobs)
        env_copy.worker_time = dict(self.worker_time)
        env_copy.worker_last_job = dict(self.worker_last_job)
        env_copy.current_time = self.current_time
        env_copy.task_assignments = dict(self.task_assignments)
        return env_copy
    
    def prioritize_ongoing_tasks(self):
        """Set ongoing tasks first in the scheduler"""
        if self.ongoing_tasks is not None:
            for job, sub_job, worker, process_time in self.ongoing_tasks:
                start_time = self.worker_time[worker]
            
                last_finished_job_time = max(
                    (end for j, _, _, _, end in self.schedule if j == job),
                    default=0
                )
                if last_finished_job_time > start_time:
                    start_time = last_finished_job_time
                        
                finish_time = start_time + process_time
                self.worker_time[worker] = finish_time
                self.worker_last_job[worker] = job
            
                self.schedule.append((job, sub_job, worker, start_time, finish_time))
                self.completed_jobs.add(sub_job)
                self.current_time = max(self.worker_time.values())

                self.task_assignments[sub_job] = worker
    
    def reset(self):
        """Reset the environment to initial state"""
        self.schedule = []
        self.completed_jobs = set()
        self.worker_time = {worker: 0 for worker in self.workers}
        self.worker_last_job = {worker: None for worker in self.workers}
        self.current_time = 0
        self.task_assignments = {}
        self.prioritize_ongoing_tasks()        
        return self.get_state()

    def previous_task_done(self, job, sub_job):
        """Check if all prerequisite tasks for this sub_job have been completed"""
        if sub_job in self.completed_jobs:
            return False

        job_sequence = self.jobs[job]
        sub_job_index = job_sequence.index(sub_job)

        if sub_job_index > 0:
            required_sub_jobs = job_sequence[:sub_job_index]
            if any(req_sub not in self.completed_jobs for req_sub in required_sub_jobs):
                return False

        return True
    
    def is_valid_action(self, job, sub_job, worker):
        """Check if an action (job-task-worker assignment) is valid"""
        # Check if job is already completed
        if sub_job in self.completed_jobs:
            return False

        # Check if worker can process this task
        if worker not in self.processing_times.get(sub_job, {}):
            return False

        # Check task dependencies/prerequisites
        job_sequence = self.jobs[job]
        sub_job_index = job_sequence.index(sub_job)

        if sub_job_index > 0:
            required_sub_jobs = job_sequence[:sub_job_index]
            if any(req_sub not in self.completed_jobs for req_sub in required_sub_jobs):
                return False
                
        # Check not_same_pic constraint
        if self.not_same_pic and sub_job in self.not_same_pic:
            prev_task = self.not_same_pic[sub_job]
            prev_worker = self.task_assignments.get(prev_task)
            if prev_worker == worker:
                return False
                
        return True

    def get_possible_actions(self):
        """Get all possible valid actions in the current state"""
        actions = [
            (job, sub_job, worker)
            for job, sub_jobs in self.jobs.items()
            for sub_job in sub_jobs
            for worker in self.workers
            if self.is_valid_action(job, sub_job, worker)
        ]
        return actions

    def step(self, action):
        """Execute an action and advance the environment state"""
        if action is None or len(action) != 3:
            return self.get_state(), -100, True
    
        job, sub_job, worker = action
        if not self.is_valid_action(job, sub_job, worker):
            return self.get_state(), -10, False
    
        process_time = self.processing_times[sub_job][worker]
        start_time = self.worker_time[worker]
    
        # Consider job sequence constraints
        last_finished_job_time = max(
            (end for j, _, _, _, end in self.schedule if j == job),
            default=0
        )
        if last_finished_job_time > start_time:
            start_time = last_finished_job_time
    
        # Apply switching penalty if worker changes jobs
        switching = self.worker_last_job[worker] is not None and self.worker_last_job[worker] != job
        if switching:
            start_time += self.switch_penalty
    
        # Calculate finish time and update state
        finish_time = start_time + process_time
        self.worker_time[worker] = finish_time
        self.worker_last_job[worker] = job
    
        self.schedule.append((job, sub_job, worker, start_time, finish_time))
        self.completed_jobs.add(sub_job)
        self.current_time = max(self.worker_time.values())
        self.task_assignments[sub_job] = worker
    
        # Compute statistics and reward
        total_jobs = sum(len(subs) for subs in self.jobs.values())
    
        last_finish_times = {
            job: max((end for j, _, _, _, end in self.schedule if j == job), default=0)
            for job in self.jobs
        }
    
        overshoot_per_job = {
            job: max(0, last_finish_times[job] - self.deadlines[job])
            for job in self.jobs
        }
    
        total_overshoot = sum(overshoot_per_job.values())
    
        # Utilization: busy time / makespan
        makespan = self.current_time
        avg_busy_time = sum(self.worker_time.values()) / len(self.worker_time)
        utilization = avg_busy_time / makespan if makespan > 0 else 0
    
        # Reward calculation
        reward = (
            -10.0 * makespan                       # minimize makespan
            - 10.0 * (1 - utilization)             # minimize idle time
            - 5.0 * total_overshoot                # penalize late jobs
            - 0.25 * (self.switch_penalty if switching else 0)  # switching penalty
        )
    
        done = len(self.completed_jobs) == total_jobs
        return self.get_state(), reward, done

    def get_state(self):
        """Get current state representation of the environment"""
        completed_vector = [
            1 if sub_job in self.completed_jobs else 0
            for job in self.jobs
            for sub_job in self.jobs[job]
        ]
        worker_loads = list(self.worker_time.values())
        return tuple(completed_vector + worker_loads)
```

```python
class GreedyAgent:
    """Greedy agent for solving FJSP"""
    
    def __init__(self, env):
        self.env = env

    def choose_action(self):
        """Choose the action that minimizes makespan"""
        actions = self.env.get_possible_actions()
        if not actions:
            return None

        # For each possible action, simulate and evaluate makespan
        results = {}
        for action in actions:
            env_copy = self.env.create_copy()
            env_copy.step(action)
            
            # Calculate makespan + deadline penalty
            makespan = env_copy.current_time
            
            # Check deadline penalties
            last_finish_times = {
                job: max((end for j, _, _, _, end in env_copy.schedule if j == job), default=0)
                for job in env_copy.jobs
            }
            
            deadline_penalties = sum(
                max(0, last_finish_times[job] - env_copy.deadlines[job])
                for job in env_copy.jobs
            )
            
            # Combined score (lower is better)
            results[action] = makespan + 0.5 * deadline_penalties
        
        return min(results, key=results.get)

    def get_best_schedule(self):
        """Generate the best schedule using greedy approach"""
        state = self.env.reset()
        done = False
        while not done:
            action = self.choose_action()
            if action is None:
                break
            state, _, done = self.env.step(action)
        return self.env.schedule, self.env.current_time
```

```python
def plot_gantt(schedule, workers, deadlines=None):
    """Plot Gantt chart of the schedule"""
    fig, ax = plt.subplots(figsize=(12, 6))

    # Assign unique colors per job
    unique_jobs = set(job for job, _, _, _, _ in schedule)
    job_colors = {job: plt.cm.tab20(i % 20) for i, job in enumerate(unique_jobs)}

    # Map workers to vertical positions
    worker_pos = {worker: i for i, worker in enumerate(workers)}

    # Draw schedule bars
    for job, sub_job, worker, start, end in schedule:
        ypos = worker_pos[worker] * 10
        ax.broken_barh([(start, end - start)], (ypos, 9), 
                        facecolors=job_colors[job], edgecolor='black')
        ax.text(start + (end - start) / 2, ypos + 4.5, f"{sub_job.split('-')[-1]}", 
                ha='center', va='center', color='black', fontsize=8)

    # Draw deadline markers if provided
    if deadlines:
        for job, deadline in deadlines.items():
            for worker_idx, worker in enumerate(workers):
                ypos = worker_idx * 10
                plt.axvline(x=deadline, ymin=ypos/100, ymax=(ypos+9)/100, 
                            color='red', linestyle='--', alpha=0.5)
            
            # Only add one label per job
            ax.text(deadline, len(workers) * 10 + 2, f"{job} Deadline", 
                    ha='center', va='bottom', color='red', rotation=90, fontsize=8)

    # Format plot
    ax.set_xlabel("Time")
    ax.set_ylabel("Workers")
    ax.set_yticks([i * 10 + 4.5 for i in range(len(workers))])
    ax.set_yticklabels(workers)
    ax.grid(True, linestyle='--', alpha=0.3)

    # Legend
    legend_patches = [mpatches.Patch(color=color, label=f"{job}") for job, color in job_colors.items()]
    ax.legend(handles=legend_patches, title="Jobs", loc="upper right")

    plt.title("Gantt Chart - Job Shop Scheduling with Deadlines")
    plt.tight_layout()
    plt.show()
```

```python
def run_greedy_scheduler(df_projects, df_pic_tasks, df_skills, ongoing_tasks, model_path):
    """Run the full greedy scheduling process"""
    # Initialize the data processor
    processor = DataProcessor()
    processor.load_catboost_model(model_path)
    
    # Process input data
    df_predict_takt, jobs, workers, not_same_pic = processor.process_data(
        df_projects, df_pic_tasks, df_skills
    )
    
    # Predict processing times
    processing_times = processor.predict_processing_times(df_predict_takt)
    
    # Calculate deadlines based on delivery dates
    deadlines = processor.calculate_delivery_based_deadlines(jobs, processing_times, df_projects)
    
    # Process ongoing tasks
    if ongoing_tasks:
        jobs, deadlines, processing_times = processor.process_ongoing_tasks(
            ongoing_tasks, jobs, deadlines, processing_times
        )
    
    # Initialize environment
    env = FJSP_Env(
        jobs=jobs,
        workers=workers,
        processing_times=processing_times,
        deadlines=deadlines,
        switch_penalty=2,
        ongoing_tasks=ongoing_tasks,
        not_same_pic=not_same_pic
    )
    
    # Create and run greedy agent
    greedy_agent = GreedyAgent(env)
    best_schedule, makespan = greedy_agent.get_best_schedule()
    
    # Check for deadline violations
    last_finish_times = {
        job: max((end for j, _, _, _, end in best_schedule if j == job), default=0)
        for job in jobs
    }
    
    deadline_violations = {
        job: finish - deadlines[job] 
        for job, finish in last_finish_times.items() 
        if finish > deadlines[job]
    }
    
    # Calculate statistics
    total_violations = sum(deadline_violations.values())
    utilization = sum(finish for _, _, _, _, finish in best_schedule) / (len(workers) * makespan) if makespan > 0 else 0
    
    # Print results
    print(f"Schedule generated with makespan: {makespan:.2f}")
    print(f"Worker utilization: {utilization:.2%}")
    print(f"Deadline violations: {len(deadline_violations)} jobs totaling {total_violations:.2f} time units")
    
    # Create result dataframe
    result_df = pd.DataFrame(
        best_schedule, 
        columns=['PN', 'Task', 'PIC', 'Start', 'Finish']
    )
    
    # Add deadline information
    result_df['Deadline'] = result_df['PN'].map(deadlines)
    result_df['Violation'] = result_df.apply(
        lambda row: max(0, row['Finish'] - row['Deadline']), axis=1
    )
    
    return result_df, best_schedule, deadlines, makespan
```
