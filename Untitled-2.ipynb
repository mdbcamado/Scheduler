```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from catboost import CatBoostRegressor
import datetime


# ================================================
# DATA PROCESSING MODULE
# ================================================

class DataProcessor:
    """Handles data loading, preprocessing and feature engineering"""
    
    def __init__(self):
        self.catboost_model = None
        
    def load_catboost_model(self, model_path):
        """Load the CatBoost model for predicting task durations"""
        self.catboost_model = CatBoostRegressor()
        self.catboost_model.load_model(model_path)
        print(f"CatBoost model loaded from {model_path}")
        
    def process_data(self, df_projects, df_pic_tasks, df_skills):
        """Process input dataframes to prepare for scheduling"""
        # Merge dataframes
        df_predict_takt = pd.merge(df_projects, df_pic_tasks, on='PRODUCT-ASSY TYPE Task', how='inner')
        df_predict_takt = df_predict_takt.merge(df_skills, on='PIC', how='inner')
        
        # Extract job structure
        jobs = df_projects.groupby('PN', sort=False)['PN Task'].apply(list).to_dict()
        
        # Get unique workers
        workers = list(df_predict_takt['PIC'].unique())
        
        # Create a mapping of tasks that shouldn't be assigned to the same worker
        not_same_pic = self._extract_not_same_pic_constraints(df_projects)
        
        return df_predict_takt, jobs, workers, not_same_pic
    
    def _extract_not_same_pic_constraints(self, df_projects):
        """Extract constraints where tasks shouldn't be assigned to the same worker"""
        not_same_pic = {}
        
        # Look for drawing tasks and final checking tasks that should have different workers
        for pn in df_projects['PN'].unique():
            drawing_tasks = df_projects[(df_projects['PN'] == pn) & 
                                       (df_projects['PN Task'].str.contains('DRAWING'))]['PN Task'].tolist()
            
            checking_tasks = df_projects[(df_projects['PN'] == pn) & 
                                        (df_projects['PN Task'].str.contains('FINAL CHECKING'))]['PN Task'].tolist()
            
            # Create constraints that drawing and checking should have different PICs
            for check_task in checking_tasks:
                for draw_task in drawing_tasks:
                    not_same_pic[check_task] = draw_task
        
        return not_same_pic
    
    def predict_processing_times(self, df_predict_takt):
        """Use CatBoost model to predict task processing times"""
        if self.catboost_model is None:
            raise ValueError("CatBoost model not loaded. Call load_catboost_model first.")
        
        # Prepare features for prediction
        cat_cols = ['Task', 'ASSY TYPE', 'PARTS TYPE', 'ORDER TYPE', 
                    'SITE_TYPE', 'PN_PREFIX', 'Designation']
        
        df_ml_dataset = df_predict_takt[[
            'Task', 'ASSY TYPE', 'PARTS TYPE',
            'ORDER TYPE', 'SITE_TYPE', 'PN_PREFIX',
            'JAGUAR', 'WCSP', 'STRIP', 'BI', 'TI', 'NON TI',
            'Designation',
            'CadExp_Enplas', 'Enplas_Years', 'CADWork_NonEnplas',
            'NonCadWork', 'Total',
            'NUM_SITE', 'NUM_STRIP'
        ]].copy()
        
        # Convert categorical columns
        for col in cat_cols:
            df_ml_dataset[col] = df_ml_dataset[col].astype('category')
        
        # Validate data for prediction
        if df_ml_dataset.empty:
            raise ValueError("Input DataFrame for prediction is empty.")

        missing_cols = set(self.catboost_model.feature_names_) - set(df_ml_dataset.columns)
        if missing_cols:
            raise ValueError(f"Missing features required by the model: {missing_cols}")
        
        # Predict processing times
        predicted_takt = self.catboost_model.predict(df_ml_dataset, prediction_type="RawFormulaVal")
        df_predict_takt['Predicted Takt'] = predicted_takt
        
        # Convert to the format needed by the scheduler
        processing_times = {}
        for task, group in df_predict_takt.groupby('PN Task'):
            processing_times[task] = dict(zip(group['PIC'], group['Predicted Takt']))
        
        return processing_times
    
    def calculate_delivery_based_deadlines(self, jobs, processing_times, df_projects):
        """Calculate deadlines based on delivery dates"""
        deadlines = {}
        today = datetime.datetime.now()
        
        for job, tasks in jobs.items():
            job_data = df_projects[df_projects['PN'] == job]
            
            if not job_data.empty and 'Prio Due' in job_data.columns and not pd.isna(job_data['Prio Due'].iloc[0]):
                # Use delivery date if available
                delivery_date = job_data['Prio Due'].iloc[0]
                
                # Ensure delivery_date is a datetime object
                if isinstance(delivery_date, (datetime.datetime, datetime.date)):
                    # Convert to time units (days * 8 hours per day)
                    days_until_delivery = (delivery_date - today).days
                    time_units = days_until_delivery * 8  # Assuming 8 time units per day
                    deadlines[job] = max(time_units, 20)  # Ensure minimum deadline
                else:
                    # If it's not a datetime, try to use it directly as a number
                    try:
                        deadlines[job] = float(delivery_date)
                    except (ValueError, TypeError):
                        # Fall back to default of 100 hours
                        deadlines[job] = 100.0
                        print(f"Using default deadline of 100 hours for {job} (could not interpret delivery date)")
            else:
                # No delivery date - use default of 100 hours
                deadlines[job] = 100.0
                print(f"Using default deadline of 100 hours for {job} (no delivery date specified)")
                
        return deadlines
    
    def _calculate_base_time(self, tasks, processing_times):
        """Helper method to calculate base time from task durations"""
        base_time = 0
        for task in tasks:
            if task in processing_times and processing_times[task]:
                min_time = min(processing_times[task].values())
                base_time += min_time
        return base_time
    
    def process_ongoing_tasks(self, ongoing_tasks, jobs, deadlines, processing_times, ongoing_deadlines=None):
        """Process ongoing tasks and update jobs, deadlines, and processing times"""
        updated_jobs = jobs.copy()
        updated_deadlines = deadlines.copy()
        updated_processing_times = processing_times.copy()
        
        for job, task, pic, takt in ongoing_tasks:
            if job in updated_jobs:
                # Add to existing job - make it the first task
                if task not in updated_jobs[job]:
                    updated_jobs[job] = [task] + updated_jobs[job]
            else:
                # Create new job with this task
                updated_jobs[job] = [task]
                
                # Set deadline from ongoing_deadlines if available, otherwise use heuristic
                if ongoing_deadlines and job in ongoing_deadlines:
                    updated_deadlines[job] = ongoing_deadlines[job]
                else:
                    # Set deadline based on product type and estimated duration as fallback
                    updated_deadlines[job] = takt * 5  # Simple heuristic: 5x task duration
            
            # Add processing time for this task/worker combination
            if task not in updated_processing_times:
                updated_processing_times[task] = {}
            updated_processing_times[task][pic] = takt
            
        return updated_jobs, updated_deadlines, updated_processing_times
        
    def extract_sheet3_deadlines(self, df_task_assignments, df_projects):
        """Extract deadlines from Sheet3 if they exist"""
        deadlines = {}
        today = datetime.datetime.now()
        
        if 'Deadline' in df_task_assignments.columns:
            # Group by PN to get unique products
            for pn in df_task_assignments['PN'].unique():
                deadline_values = df_task_assignments[df_task_assignments['PN'] == pn]['Deadline']
                
                # Try to get the first non-NaN value
                valid_deadlines = deadline_values.dropna()
                
                if not valid_deadlines.empty:
                    deadline_value = valid_deadlines.iloc[0]
                    
                    # Check if deadline is a date
                    if isinstance(deadline_value, (datetime.datetime, datetime.date)):
                        # Convert date to time units (days * 8 hours)
                        days_until_deadline = (deadline_value - today).days
                        time_units = days_until_deadline * 8  # 8 hours per day
                        deadlines[pn] = max(time_units, 10)  # Ensure minimum deadline
                    else:
                        # Use the value directly if it's not a date
                        try:
                            deadlines[pn] = float(deadline_value)
                        except (ValueError, TypeError):
                            # Use default deadline of 100 hours
                            deadlines[pn] = 100.0
                            print(f"Using default deadline of 100 hours for {pn} in Sheet3 (invalid value)")
                else:
                    # Use default deadline of 100 hours
                    deadlines[pn] = 100.0
                    print(f"Using default deadline of 100 hours for {pn} in Sheet3 (no deadline specified)")
                    
        return deadlines


# ================================================
# ENVIRONMENT MODULE
# ================================================

class FJSP_Env:
    """Flexible Job Shop Scheduling Environment"""
    
    def __init__(self, jobs, workers, processing_times, deadlines, 
                 switch_penalty=2, ongoing_tasks=None,
                 not_same_pic=None):
        self.jobs = jobs
        self.workers = workers
        self.processing_times = processing_times
        self.deadlines = deadlines
        self.switch_penalty = switch_penalty
        self.ongoing_tasks = ongoing_tasks
        self.not_same_pic = not_same_pic
        self.reset()

    def create_copy(self):
        """Create a deep copy of the environment for simulation"""
        env_copy = FJSP_Env(
            jobs=self.jobs,
            workers=self.workers,
            processing_times=self.processing_times,
            deadlines=self.deadlines,
            switch_penalty=self.switch_penalty,
            ongoing_tasks=self.ongoing_tasks,
            not_same_pic=self.not_same_pic
        )
        env_copy.schedule = list(self.schedule)
        env_copy.completed_jobs = set(self.completed_jobs)
        env_copy.worker_time = dict(self.worker_time)
        env_copy.worker_last_job = dict(self.worker_last_job)
        env_copy.current_time = self.current_time
        env_copy.task_assignments = dict(self.task_assignments)
        return env_copy
    
    def prioritize_ongoing_tasks(self):
        """Set ongoing tasks first in the scheduler"""
        if self.ongoing_tasks is not None:
            for job, sub_job, worker, process_time in self.ongoing_tasks:
                start_time = self.worker_time[worker]
            
                last_finished_job_time = max(
                    (end for j, _, _, _, end in self.schedule if j == job),
                    default=0
                )
                if last_finished_job_time > start_time:
                    start_time = last_finished_job_time
                        
                finish_time = start_time + process_time
                self.worker_time[worker] = finish_time
                self.worker_last_job[worker] = job
            
                self.schedule.append((job, sub_job, worker, start_time, finish_time))
                self.completed_jobs.add(sub_job)
                self.current_time = max(self.worker_time.values())

                self.task_assignments[sub_job] = worker
    
    def reset(self):
        """Reset the environment to initial state"""
        self.schedule = []
        self.completed_jobs = set()
        self.worker_time = {worker: 0 for worker in self.workers}
        self.worker_last_job = {worker: None for worker in self.workers}
        self.current_time = 0
        self.task_assignments = {}
        self.prioritize_ongoing_tasks()        
        return self.get_state()

    def previous_task_done(self, job, sub_job):
        """Check if all prerequisite tasks for this sub_job have been completed"""
        if sub_job in self.completed_jobs:
            return False

        job_sequence = self.jobs[job]
        sub_job_index = job_sequence.index(sub_job)

        if sub_job_index > 0:
            required_sub_jobs = job_sequence[:sub_job_index]
            if any(req_sub not in self.completed_jobs for req_sub in required_sub_jobs):
                return False

        return True
    
    def get_tasks_for_assignment(self):
        """Get tasks that are currently ready to be assigned"""
        available_tasks = []
        for job, sub_jobs in self.jobs.items():
            for sub_job in sub_jobs:
                if sub_job not in self.completed_jobs and self.previous_task_done(job, sub_job):
                    available_tasks.append((job, sub_job))
        return available_tasks
    
    def is_valid_action(self, job, sub_job, worker):
        """Check if an action (job-task-worker assignment) is valid"""
        # Check if job is already completed
        if sub_job in self.completed_jobs:
            return False

        # Check if worker can process this task
        if worker not in self.processing_times.get(sub_job, {}):
            return False

        # Check task dependencies/prerequisites
        job_sequence = self.jobs[job]
        sub_job_index = job_sequence.index(sub_job)

        if sub_job_index > 0:
            required_sub_jobs = job_sequence[:sub_job_index]
            if any(req_sub not in self.completed_jobs for req_sub in required_sub_jobs):
                return False
                
        # Check not_same_pic constraint
        if self.not_same_pic and sub_job in self.not_same_pic:
            prev_task = self.not_same_pic[sub_job]
            prev_worker = self.task_assignments.get(prev_task)
            if prev_worker == worker:
                return False
                
        return True

    def get_possible_actions(self):
        """Get all possible valid actions in the current state"""
        actions = [
            (job, sub_job, worker)
            for job, sub_jobs in self.jobs.items()
            for sub_job in sub_jobs
            for worker in self.workers
            if self.is_valid_action(job, sub_job, worker)
        ]
        return actions

    def step(self, action):
        """Execute an action and advance the environment state"""
        if action is None or len(action) != 3:
            return self.get_state(), -100, True
    
        job, sub_job, worker = action
        if not self.is_valid_action(job, sub_job, worker):
            return self.get_state(), -10, False
    
        process_time = self.processing_times[sub_job][worker]
        start_time = self.worker_time[worker]
    
        # Consider job sequence constraints
        last_finished_job_time = max(
            (end for j, _, _, _, end in self.schedule if j == job),
            default=0
        )
        if last_finished_job_time > start_time:
            start_time = last_finished_job_time
    
        # Apply switching penalty if worker changes jobs
        switching = self.worker_last_job[worker] is not None and self.worker_last_job[worker] != job
        if switching:
            start_time += self.switch_penalty
    
        # Calculate finish time and update state
        finish_time = start_time + process_time
        self.worker_time[worker] = finish_time
        self.worker_last_job[worker] = job
    
        self.schedule.append((job, sub_job, worker, start_time, finish_time))
        self.completed_jobs.add(sub_job)
        self.current_time = max(self.worker_time.values())
        self.task_assignments[sub_job] = worker
    
        # Compute statistics and reward
        total_jobs = sum(len(subs) for subs in self.jobs.values())
    
        last_finish_times = {
            job: max((end for j, _, _, _, end in self.schedule if j == job), default=0)
            for job in self.jobs
        }
    
        overshoot_per_job = {
            job: max(0, last_finish_times[job] - self.deadlines[job])
            for job in self.jobs
        }
    
        total_overshoot = sum(overshoot_per_job.values())
    
        # Utilization: busy time / makespan
        makespan = self.current_time
        avg_busy_time = sum(self.worker_time.values()) / len(self.worker_time)
        utilization = avg_busy_time / makespan if makespan > 0 else 0
    
        # Reward calculation
        reward = (
            -10.0 * makespan                       # minimize makespan
            - 10.0 * (1 - utilization)             # minimize idle time
            - 5.0 * total_overshoot                # penalize late jobs
            - 0.25 * (self.switch_penalty if switching else 0)  # switching penalty
        )
    
        done = len(self.completed_jobs) == total_jobs
        return self.get_state(), reward, done

    def get_state(self):
        """Get current state representation of the environment"""
        completed_vector = [
            1 if sub_job in self.completed_jobs else 0
            for job in self.jobs
            for sub_job in self.jobs[job]
        ]
        worker_loads = list(self.worker_time.values())
        return tuple(completed_vector + worker_loads)


# ================================================
# AGENT MODULE
# ================================================

class GreedyAgent:
    """Greedy agent that uses one-step lookahead to minimize makespan"""
    
    def __init__(self, env):
        self.env = env
        
    def choose_action(self):
        """Choose the best action using simulation-based lookahead"""
        actions = self.env.get_possible_actions()
        if not actions:
            return None

        # Simulate each possible action and see which one results in the minimum current_time
        results = {}
        for action in actions:
            # Create a copy of the environment to simulate this action
            env_copy = self.env.create_copy()
            # Take the action in the copied environment
            env_copy.step(action)
            # Store the resulting current_time
            results[action] = env_copy.current_time
        
        # Choose the action that results in the minimum current_time
        return min(results, key=results.get)   

    def get_best_schedule(self):
        """Generate the full schedule using lookahead greedy algorithm"""
        state = self.env.reset()
        done = False
        
        while not done:
            action = self.choose_action()
            if action is None:
                break
                
            state, _, done = self.env.step(action)
            
        return self.env.schedule, self.env.current_time


class HumanAgent:
    """Agent that follows human-assigned task allocation"""
    
    def __init__(self, env, task_assignments):
        self.env = env
        self.task_assignments = task_assignments

    def get_schedule(self):
        """Generate schedule based on human task assignments"""
        state = self.env.reset()
        done = False

        while not done:
            tasks_for_assignment = self.env.get_tasks_for_assignment()
            if not tasks_for_assignment:
                break

            for job, sub_job in tasks_for_assignment:
                if sub_job not in self.task_assignments:
                    print(f'No worker assigned for task {sub_job}')
                    done = True
                    break

                assigned_pic = self.task_assignments[sub_job]

                if sub_job not in self.env.processing_times or assigned_pic not in self.env.processing_times[sub_job]:
                    print(f'No processing time set for worker {assigned_pic} for task {sub_job}.')
                    done = True
                    break

                if hasattr(self.env, "not_same_pic") and sub_job in self.env.not_same_pic:
                    prev_task = self.env.not_same_pic[sub_job]
                    prev_worker = self.task_assignments.get(prev_task)
                    if prev_worker == assigned_pic:
                        print(f'Same worker {assigned_pic} assigned for {prev_task} and {sub_job}.')
                        done = True
                        break

                state, _, done = self.env.step((job, sub_job, assigned_pic))

        return self.env.schedule, self.env.current_time


# ================================================
# VISUALIZATION MODULE
# ================================================

def plot_gantt(schedule, workers, deadlines=None, title=None):
    """Plot Gantt chart of the schedule"""
    fig, ax = plt.subplots(figsize=(14, 8))  # Larger figure size for better visibility

    # Assign unique colors per job
    unique_jobs = set(job for job, _, _, _, _ in schedule)
    job_colors = {job: plt.cm.tab20(i % 20) for i, job in enumerate(unique_jobs)}

    # Map workers to vertical positions
    worker_pos = {worker: i for i, worker in enumerate(workers)}

    # Calculate the makespan to adjust the x-axis
    makespan = max((end for _, _, _, _, end in schedule), default=0)
    
    # Draw schedule bars
    for job, sub_job, worker, start, end in schedule:
        ypos = worker_pos[worker] * 10
        bar_height = 8  # Slightly smaller than before to add spacing
        ax.broken_barh([(start, end - start)], (ypos, bar_height), 
                       facecolors=job_colors[job], edgecolor='black')
        
        # Adjust text size based on bar width
        bar_width = end - start
        fontsize = min(9, max(6, bar_width / 5))  # Adaptive font size
        
        # Only add text if bar is wide enough
        if bar_width > 5:
            ax.text(start + (end - start) / 2, ypos + bar_height/2, f"{sub_job.split('-')[-1]}", 
                   ha='center', va='center', color='black', fontsize=fontsize)

    # Format plot
    ax.set_xlabel("Time (hours)")
    ax.set_ylabel("Workers")
    ax.set_yticks([i * 10 + 4 for i in range(len(workers))])
    ax.set_yticklabels(workers)
    
    # Set x-axis limits with some padding
    x_padding = max(20, makespan * 0.05)  # Add 5% padding or minimum 20 units
    ax.set_xlim(-5, makespan + x_padding)
    
    # Add grid for better readability
    ax.grid(True, linestyle='--', alpha=0.3)
    
    # Add hourly markers
    x_interval = max(8, makespan // 20)  # Adaptive interval based on makespan
    ax.set_xticks(range(0, int(makespan + x_padding), x_interval))
    
    # Add worker utilization markers
    for worker in workers:
        worker_jobs = [(job, start, end) for job, _, w, start, end in schedule if w == worker]
        if worker_jobs:
            total_time = sum(end - start for _, start, end in worker_jobs)
            worker_makespan = max(end for _, _, end in worker_jobs)
            utilization = total_time / worker_makespan if worker_makespan > 0 else 0
            ypos = worker_pos[worker] * 10 - 0.5
            ax.text(makespan + 5, ypos + 4, f"{utilization:.1%} util", 
                   ha='left', va='center', fontsize=8)

    # Legend
    legend_patches = [mpatches.Patch(color=color, label=f"{job}") for job, color in job_colors.items()]
    ax.legend(handles=legend_patches, title="Jobs", loc="upper right")

    # Title
    plot_title = "Gantt Chart - Job Shop Scheduling"
    if title:
        plot_title = f"{plot_title} - {title}"
    plt.title(plot_title)
    
    plt.tight_layout()
    plt.show()


# ================================================
# UTILITY FUNCTIONS MODULE
# ================================================

def load_ongoing_tasks(excel_filename):
    """Load ongoing tasks from Sheet4 of the Excel file"""
    try:
        df_ongoing = pd.read_excel(excel_filename, sheet_name="Sheet4")
        if df_ongoing.empty:
            print("No ongoing tasks found in Sheet4.")
            return None, {}
            
        # Check required columns
        required_columns = ['PN', 'PN Task', 'PIC', 'Duration']
        missing_columns = [col for col in required_columns if col not in df_ongoing.columns]
        
        if missing_columns:
            print(f"Warning: Missing columns in Sheet4: {missing_columns}")
            print("Using default format for ongoing tasks.")
            return None, {}
            
        # Convert to list of lists format
        ongoing_tasks = []
        ongoing_deadlines = {}
        today = datetime.datetime.now()
        
        for _, row in df_ongoing.iterrows():
            # Extract task info
            ongoing_tasks.append([
                row['PN'],
                row['PN Task'],
                row['PIC'],
                row['Duration']
            ])
            
            # Always set a deadline for each task - use provided value or default of 100
            if 'Deadline' in df_ongoing.columns and not pd.isna(row['Deadline']):
                # Check if the deadline is a date
                if isinstance(row['Deadline'], (datetime.datetime, datetime.date)):
                    # Convert date to time units (days * 8 hours)
                    days_until_deadline = (row['Deadline'] - today).days
                    time_units = days_until_deadline * 8
                    ongoing_deadlines[row['PN']] = max(time_units, 10)  # Ensure minimum deadline
                else:
                    # Use the value directly as time units
                    ongoing_deadlines[row['PN']] = float(row['Deadline'])
            else:
                # Use default deadline of 100 hours if not specified
                ongoing_deadlines[row['PN']] = 100.0
                print(f"Using default deadline of 100 hours for {row['PN']}")
                
        return ongoing_tasks, ongoing_deadlines
    except Exception as e:
        print(f"Error loading ongoing tasks: {e}")
        return None, {}

# ================================================
# SCHEDULER RUNNER MODULE
# ================================================

def setup_scheduling_environment(df_projects, df_pic_tasks, df_skills, ongoing_tasks, ongoing_deadlines, model_path):
    """Set up the scheduling environment with common components"""
    # Initialize the data processor
    processor = DataProcessor()
    processor.load_catboost_model(model_path)
    
    # Process input data
    df_predict_takt, jobs, workers, not_same_pic = processor.process_data(
        df_projects, df_pic_tasks, df_skills
    )
    
    # Predict processing times
    processing_times = processor.predict_processing_times(df_predict_takt)
    
    # Calculate deadlines based on delivery dates
    deadlines = processor.calculate_delivery_based_deadlines(jobs, processing_times, df_projects)
    
    # Process ongoing tasks
    if ongoing_tasks:
        jobs, deadlines, processing_times = processor.process_ongoing_tasks(
            ongoing_tasks, jobs, deadlines, processing_times, ongoing_deadlines
        )
    
    return processor, jobs, workers, processing_times, deadlines, not_same_pic


def create_result_dataframe(schedule, jobs, deadlines):
    """Create a DataFrame from the schedule results"""
    # Check for deadline violations
    last_finish_times = {
        job: max((end for j, _, _, _, end in schedule if j == job), default=0)
        for job in jobs
    }
    
    deadline_violations = {
        job: finish - deadlines[job] 
        for job, finish in last_finish_times.items() 
        if finish > deadlines[job]
    }
    
    # Create result dataframe
    result_df = pd.DataFrame(
        schedule, 
        columns=['PN', 'Task', 'PIC', 'Start', 'Finish']
    )
    
    # Add deadline information
    result_df['Deadline'] = result_df['PN'].map(deadlines)
    result_df['Violation'] = result_df.apply(
        lambda row: max(0, row['Finish'] - row['Deadline']), axis=1
    )
    
    return result_df


def print_schedule_stats(result_df, makespan, workers):
    """Print statistics about the schedule"""
    # Calculate statistics
    deadline_violations = result_df[result_df['Violation'] > 0]
    total_violations = deadline_violations['Violation'].sum()
    
    # Calculate worker utilization
    total_work_time = result_df['Finish'].max() - result_df['Start'].min()
    total_worker_time = len(workers) * total_work_time if total_work_time > 0 else 1
    busy_time = sum(result_df['Finish'] - result_df['Start'])
    utilization = busy_time / total_worker_time
    
    # Print results
    print(f"Schedule generated with makespan: {makespan:.2f}")
    print(f"Worker utilization: {utilization:.2%}")
    print(f"Deadline violations: {len(deadline_violations)} tasks totaling {total_violations:.2f} time units")


# ================================================
# MAIN EXECUTION MODULE
# ================================================

def run_scheduling_comparison(excel_filename, skills_filename, model_path):
    """Run both scheduler types and compare results"""
    print("Loading data...")
    
    # Load data from Excel
    df_projects = pd.read_excel(excel_filename, sheet_name="Sheet1")
    df_pic_tasks = pd.read_excel(excel_filename, sheet_name="Sheet2")
    df_task_assignments = pd.read_excel(excel_filename, sheet_name="Sheet3")
    df_skills = pd.read_excel(skills_filename)
    
    # Load ongoing tasks from Sheet4
    ongoing_tasks, ongoing_deadlines = load_ongoing_tasks(excel_filename)
    
    if ongoing_tasks:
        print(f"Loaded {len(ongoing_tasks)} ongoing tasks from Sheet4.")
    else:
        print("No ongoing tasks found in Sheet4 or unable to load them.")
        ongoing_tasks = None
        ongoing_deadlines = {}
    
    # ===== GREEDY SCHEDULER =====
    print("\n" + "="*50)
    print("RUNNING GREEDY SCHEDULER")
    print("="*50)
    
    # Create a fresh environment for the greedy scheduler
    processor_greedy, jobs_greedy, workers_greedy, processing_times_greedy, deadlines_greedy, not_same_pic_greedy = setup_scheduling_environment(
        df_projects.copy(), df_pic_tasks.copy(), df_skills.copy(), 
        ongoing_tasks.copy() if ongoing_tasks else None, 
        ongoing_deadlines.copy(), model_path
    )
    
    # Initialize environment
    env_greedy = FJSP_Env(
        jobs=jobs_greedy,
        workers=workers_greedy,
        processing_times=processing_times_greedy,
        deadlines=deadlines_greedy,
        switch_penalty=2,
        ongoing_tasks=ongoing_tasks.copy() if ongoing_tasks else None,
        not_same_pic=not_same_pic_greedy
    )
    
    # Create and run greedy agent
    print("Running greedy agent with one-step lookahead simulation...")
    greedy_agent = GreedyAgent(env_greedy)
    schedule_greedy, makespan_greedy = greedy_agent.get_best_schedule()
    
    # Get stats and create result dataframe
    result_greedy = create_result_dataframe(schedule_greedy, jobs_greedy, deadlines_greedy)
    print_schedule_stats(result_greedy, makespan_greedy, workers_greedy)
    
    # ===== HUMAN SCHEDULER =====
    print("\n" + "="*50)
    print("RUNNING HUMAN SCHEDULER")
    print("="*50)
    
    # Create a fresh environment for the human scheduler
    processor_human, jobs_human, workers_human, processing_times_human, deadlines_human, not_same_pic_human = setup_scheduling_environment(
        df_projects.copy(), df_pic_tasks.copy(), df_skills.copy(), 
        ongoing_tasks.copy() if ongoing_tasks else None, 
        ongoing_deadlines.copy(), model_path
    )
    
    # Check if Sheet3 has deadlines and update them if so
    sheet3_deadlines = processor_human.extract_sheet3_deadlines(df_task_assignments, df_projects)
    if sheet3_deadlines:
        for job, deadline in sheet3_deadlines.items():
            deadlines_human[job] = deadline
    
    # Initialize environment
    env_human = FJSP_Env(
        jobs=jobs_human,
        workers=workers_human,
        processing_times=processing_times_human,
        deadlines=deadlines_human,
        switch_penalty=2,
        ongoing_tasks=ongoing_tasks.copy() if ongoing_tasks else None,
        not_same_pic=not_same_pic_human
    )
    
    # Extract task assignments
    task_assignments = df_task_assignments.set_index('PN Task')['PIC'].to_dict()
    
    # Create and run human agent
    human_agent = HumanAgent(env_human, task_assignments)
    schedule_human, makespan_human = human_agent.get_schedule()
    
    # Get stats and create result dataframe
    result_human = create_result_dataframe(schedule_human, jobs_human, deadlines_human)
    print_schedule_stats(result_human, makespan_human, workers_human)
    
    # Save results to Excel
    result_greedy.to_excel('schedule_greedy_results.xlsx', index=False)
    result_human.to_excel('schedule_human_results.xlsx', index=False)
    
    # Visualize schedules
    print("\n" + "="*50)
    print("VISUALIZING GREEDY SCHEDULE")
    print("="*50)
    plot_gantt(schedule_greedy, workers_greedy, deadlines_greedy, "Greedy Algorithm")
    
    print("\n" + "="*50)
    print("VISUALIZING HUMAN SCHEDULE")
    print("="*50)
    plot_gantt(schedule_human, workers_human, deadlines_human, "Human Assignments")
    
    # Compare results
    print("\n" + "="*50)
    print("COMPARING SCHEDULES")
    print("="*50)
    print(f"Greedy Scheduler Makespan: {makespan_greedy:.2f}")
    print(f"Human Scheduler Makespan: {makespan_human:.2f}")
    print(f"Difference: {makespan_human - makespan_greedy:.2f} ({(makespan_human - makespan_greedy) / makespan_greedy * 100:.2f}%)")
    
    # Calculate violation statistics
    greedy_violations = result_greedy[result_greedy['Violation'] > 0]['Violation'].sum()
    human_violations = result_human[result_human['Violation'] > 0]['Violation'].sum()
    
    print(f"Greedy Scheduler Total Deadline Violations: {greedy_violations:.2f}")
    print(f"Human Scheduler Total Deadline Violations: {human_violations:.2f}")
    
    return result_greedy, result_human


if __name__ == "__main__":
    # Example execution:
    excel_filename = "sample_schedule.xlsx"
    skills_filename = "PIC_skills_edited.xlsx"
    model_path = "catboost_model.cbm"
    
    # Run scheduling comparison - ongoing tasks are now loaded from Sheet4
    result_greedy, result_human = run_scheduling_comparison(
        excel_filename, skills_filename, model_path
    )
```

```python
import pandas as pd
import re

def process_core_data(df_core):
    """
    Process a DataFrame with three core columns (PN, PRODUCT-ASSY TYPE, Task)
    and generate all derivative columns based on the patterns observed in the data.
    
    Parameters:
    df_core (pd.DataFrame): DataFrame with columns ['PN', 'PRODUCT-ASSY TYPE', 'Task']
    
    Returns:
    pd.DataFrame: Complete DataFrame with all derivative columns
    """
    
    # Create a copy to avoid modifying the original
    df = df_core.copy()
    
    # Generate derivative columns
    df['PN Task'] = df['PN'] + '-' + df['Task']
    df['PRODUCT-ASSY TYPE Task'] = df['PRODUCT-ASSY TYPE'] + '-' + df['Task']
    
    # Extract ASSY TYPE from PRODUCT-ASSY TYPE
    def extract_assy_type(product_assy_type):
        # Pattern: ASSY_TYPE + PARTS_TYPE + ORDER_TYPE
        # Updated with all discovered patterns
        
        # Define known patterns (order matters - check longer ones first)
        assy_types = ['JAGUAR', 'BGAT', 'STRIP', 'QFN', 'FP', 'OTQ', 'LGAT', 'WCSP']
        parts_types = ['SOCKET & LID', 'SOCKET', 'LID', 'PART']  # Check longer patterns first
        order_types = [
            'DIFFICULTINQUIRY', 'DIFFICULTPO', 'EASYINQUIRY', 'EASYPO',
            'MODERATEINQUIRY', 'MODERATEPO', 'NEWINQUIRY', 'NEWPO',
            'W/ REFINQUIRY', 'W/ REFPO'
        ]
        
        for assy in assy_types:
            if product_assy_type.startswith(assy):
                remaining = product_assy_type[len(assy):]
                
                # Handle "SOCKET & LID" case specially (it has spaces)
                if ' & LID' in remaining:
                    # Find where SOCKET ends and & LID begins
                    socket_lid_pos = remaining.find(' & LID')
                    if socket_lid_pos >= 0 and remaining.startswith('SOCKET'):
                        parts = 'SOCKET & LID'
                        order = remaining[socket_lid_pos + len(' & LID'):]
                        return assy, parts, order
                
                # Find other parts types (no spaces)
                for parts in ['SOCKET', 'LID', 'PART']:
                    if remaining.startswith(parts):
                        order = remaining[len(parts):]
                        return assy, parts, order
        
        return None, None, None
    
    df[['ASSY TYPE', 'PARTS TYPE', 'ORDER TYPE']] = df['PRODUCT-ASSY TYPE'].apply(
        lambda x: pd.Series(extract_assy_type(x))
    )
    
    # Generate INQUIRY_OR_PO and Order Type based on ORDER TYPE
    def get_inquiry_po(order_type):
        if order_type and ('INQUIRY' in order_type):
            return 'INQUIRY', 'INQUIRY'
        else:
            return 'PO', 'PO'
    
    df[['INQUIRY_OR_PO', 'Order Type']] = df['ORDER TYPE'].apply(
        lambda x: pd.Series(get_inquiry_po(x))
    )
    
    # Extract NUM_SITE, NUM_STRIP, SITE_TYPE, PN_PREFIX from PN
    def extract_pn_components(pn):
        # Pattern examples:
        # QFN-28X32EPK-0.5-001 -> PREFIX=QFN, SITE=28, STRIP=32, TYPE=EPK
        # BGAT-113EP-0.5-001-00 -> PREFIX=BGAT, SITE=113, STRIP=1, TYPE=EP
        # FP-28X35EPK-0.65-001-C0 -> PREFIX=FP, SITE=28, STRIP=35, TYPE=EPK
        
        parts = pn.split('-')
        prefix = parts[0]
        
        # Extract dimensions and type from second part
        if 'X' in parts[1]:
            # Format like 28X32EPK or 16X40EP
            dim_type = parts[1]
            if 'EPK' in dim_type:
                dims = dim_type.replace('EPK', '')
                site_type = 'EPK'
            elif 'EP' in dim_type:
                dims = dim_type.replace('EP', '')
                site_type = 'EP'
            else:
                dims = dim_type
                site_type = ''
            
            if 'X' in dims:
                site, strip = dims.split('X')
                site = int(site)
                strip = int(strip)
            else:
                site, strip = int(dims), 1
        else:
            # Format like 113EP or 48EPK
            dim_type = parts[1]
            if 'EPK' in dim_type:
                site = int(dim_type.replace('EPK', ''))
                site_type = 'EPK'
            elif 'EP' in dim_type:
                site = int(dim_type.replace('EP', ''))
                site_type = 'EP'
            else:
                site = int(dim_type)
                site_type = ''
            strip = 1
        
        return site, strip, site_type, prefix
    
    df[['NUM_SITE', 'NUM_STRIP', 'SITE_TYPE', 'PN_PREFIX']] = df['PN'].apply(
        lambda x: pd.Series(extract_pn_components(x))
    )
    
    # Reorder columns to match the original structure
    column_order = [
        'project_id', 'Prio Due', 'Recommit Date', 
        'PN', 'PRODUCT-ASSY TYPE', 'Task', 'PN Task', 'PRODUCT-ASSY TYPE Task',
        'ASSY TYPE', 'PARTS TYPE', 'ORDER TYPE', 'INQUIRY_OR_PO', 'Order Type',
        'NUM_SITE', 'NUM_STRIP', 'SITE_TYPE', 'PN_PREFIX', 'project_id'
    ]
    
    return df[column_order]
```

```python
# Function to load data from file and process it
def load_and_process_from_file(file_path, core_columns_only=True):
    """
    Load data from file and process it
    
    Parameters:
    file_path (str): Path to the input file
    core_columns_only (bool): If True, only use the three core columns
    
    Returns:
    pd.DataFrame: Processed DataFrame
    """
    # Read the file
    df = pd.read_csv(file_path, sep='\t')
    
    if core_columns_only:
        # Extract only the core columns
        core_columns = ['PN', 'PRODUCT-ASSY TYPE', 'Task']
        df_core = df[core_columns].copy()
        
        # Process to generate derivative columns
        return process_core_data(df_core)
    else:
        return df

if __name__ == "__main__":
    # Run the example
    result_df = load_and_process_from_file
    
    # Save to file if needed
    # result_df.to_csv('processed_data.csv', index=False)
    # result_df.to_excel('processed_data.xlsx', index=False)
```

```python
import pandas as pd
import smartsheet
import logging

# Configure basic logging
logging.basicConfig(level=logging.INFO, 
                    format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

def get_smartsheet_data(sheet_id: str, api_token: str) -> pd.DataFrame:
    """
    Retrieve data from a Smartsheet and convert to DataFrame
    
    Args:
        sheet_id (str): Unique identifier for the Smartsheet
        api_token (str): Smartsheet API token
    
    Returns:
        pd.DataFrame: DataFrame with sheet data
    """
    try:
        # Initialize client with API access token
        smart = smartsheet.Smartsheet(api_token)
        
        # Load the sheet
        sheet = smart.Sheets.get_sheet(sheet_id)
        
        # Extract column names
        columns = [col.title for col in sheet.columns]
        
        # Process rows
        rows = []
        for row in sheet.rows:
            cells = [
                cell.value if hasattr(cell, 'value') and cell.value is not None 
                else None 
                for cell in row.cells
            ]
            
            # Only append non-empty rows
            if any(cell is not None for cell in cells):
                rows.append(cells)
        
        # Create DataFrame
        df = pd.DataFrame(rows, columns=columns)
        
        logger.info(f"Successfully retrieved {len(df)} rows from sheet {sheet_id}")
        return df
    
    except Exception as e:
        logger.error(f"Error retrieving sheet {sheet_id}: {e}")
        raise
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load your data
df = pd.read_csv("data1.csv", parse_dates=['contract_end'])

# Convert dph_value to numeric and drop NA
df['dph_value'] = pd.to_numeric(df['dph_value'], errors='coerce')
df = df.dropna(subset=['dph_value'])

# Preprocess contract_type: merge categories as requested
def map_contract_type(ct):
    if ct in ['MCLOO', 'LOO']:
        return 'LOO'
    elif ct in ['OO', 'MCOO']:
        return 'OO'
    else:
        return ct

df['contract_type'] = df['contract_type'].apply(map_contract_type)

# Group by termination wave (contract_end date)
grouped = df.groupby('contract_end').agg(
    group_size=('contract_id', 'count'),
    avg_dph=('dph_value', 'mean'),
    std_dph=('dph_value', 'std'),
    min_dph=('dph_value', 'min'),
    max_dph=('dph_value', 'max'),
    contract_type=('contract_type', lambda x: x.mode()[0])  # most common contract_type in group
).reset_index()

# Fill NaNs in std_dph (e.g. when group_size=1)
grouped['std_dph'] = grouped['std_dph'].fillna(0)

# One-hot encode the contract_type (categorical)
encoder = OneHotEncoder(sparse_output=False)
contract_type_encoded = encoder.fit_transform(grouped[['contract_type']])
contract_type_cols = encoder.get_feature_names_out(['contract_type'])
contract_type_df = pd.DataFrame(contract_type_encoded, columns=contract_type_cols, index=grouped.index)

# Combine features for clustering
features = ['group_size', 'avg_dph', 'std_dph', 'min_dph', 'max_dph']
X = pd.concat([grouped[features], contract_type_df], axis=1)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
grouped['cluster'] = kmeans.fit_predict(X_scaled)

# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
grouped['PC1'] = X_pca[:, 0]
grouped['PC2'] = X_pca[:, 1]

# Plot clusters in PCA space
plt.figure(figsize=(10, 6))
sns.scatterplot(data=grouped, x='PC1', y='PC2', hue='cluster', palette='Set2', s=100)
plt.title('Termination Waves Clustering with Contract Type Included')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()

# Optional: Print cluster samples
for c in sorted(grouped['cluster'].unique()):
    print(f"\nCluster {c} sample:")
    print(grouped[grouped['cluster'] == c][['contract_end', 'group_size', 'avg_dph', 'contract_type']].head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load the data
df = pd.read_csv("data1.csv", parse_dates=['contract_end'])

# Clean dph_value column
df['dph_value'] = pd.to_numeric(df['dph_value'], errors='coerce')
df = df.dropna(subset=['dph_value'])

# Bundle contract types
df['contract_type'] = df['contract_type'].replace({
    'MCLOO': 'LOO',
    'LOO': 'LOO',
    'MCOO': 'OO',
    'OO': 'OO'
})

# Group by contract_end
grouped = df.groupby(['contract_end']).agg(
    group_size=('contract_id', 'count'),
    avg_dph=('dph_value', 'mean'),
    std_dph=('dph_value', 'std'),
    min_dph=('dph_value', 'min'),
    max_dph=('dph_value', 'max'),
    contract_type=('contract_type', lambda x: x.mode()[0] if not x.mode().empty else 'Unknown')
).reset_index()

# Fill NaN in std_dph
grouped['std_dph'] = grouped['std_dph'].fillna(0)

# Clustering (excluding contract_type)
features = ['group_size', 'avg_dph', 'std_dph', 'min_dph', 'max_dph']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(grouped[features])

kmeans = KMeans(n_clusters=3, random_state=42)
grouped['cluster'] = kmeans.fit_predict(X_scaled)

# PCA for visualization
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
grouped['PC1'] = X_pca[:, 0]
grouped['PC2'] = X_pca[:, 1]

# PCA scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=grouped, x='PC1', y='PC2', hue='cluster', palette='Set2', s=100)
plt.title('Clusters of Termination Waves (without contract_type)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()

# Cluster samples
for c in sorted(grouped['cluster'].unique()):
    print(f"\nCluster {c} sample:")
    print(grouped[grouped['cluster'] == c][['contract_end', 'group_size', 'avg_dph', 'contract_type']].head())

# Contract type distribution by cluster
contract_type_dist = grouped.groupby(['cluster', 'contract_type']).size().unstack(fill_value=0)
print("\nContract type distribution by cluster:")
print(contract_type_dist)

# Contract type stacked bar chart
contract_type_dist.plot(kind='bar', stacked=True, colormap='Set2', figsize=(10, 6))
plt.title('Contract Type Distribution by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.legend(title='Contract Type')
plt.tight_layout()
plt.show()

#  Box plots for each feature across clusters
features_to_plot = ['group_size', 'avg_dph', 'std_dph', 'min_dph', 'max_dph']
for feature in features_to_plot:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=grouped, x='cluster', y=feature, palette='Set2')
    plt.title(f'Distribution of {feature} by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel(feature)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataframe
df = pd.read_csv("data1.csv")  # Modify path as needed

# Clean column names
df.columns = df.columns.str.strip().str.lower()

# Parse date and clean numeric column
df['contract_end'] = pd.to_datetime(df['contract_end'], errors='coerce')
df['dph_value'] = pd.to_numeric(df['dph_value'], errors='coerce')
df = df.dropna(subset=['contract_end', 'dph_value_minus_10'])

# Create workweek bin (preserving year)
df['workweek'] = df['contract_end'].dt.strftime('%Y-W%U')  # %U gives week number (Sun-start)
# Alternative: use %V for ISO weeks (Mon-start), adjust depending on your preference
# df['workweek'] = df['contract_end'].dt.strftime('%G-W%V')  # for ISO week

# Group by workweek
agg_df = df.groupby('workweek').agg(
    termination_count=('driver_id', 'count'),
    avg_dph=('dph_value', 'mean')
).reset_index()

# Compute correlation
correlation = agg_df['termination_count'].corr(agg_df['avg_dph'])
print(f"Correlation between weekly termination count and average DPH: {correlation:.2f}")

# Plot
plt.figure(figsize=(10, 6))
sns.regplot(data=agg_df, x='termination_count', y='avg_dph', ci=None)
plt.title(f"Weekly Avg DPH vs Termination Count (Correlation = {correlation:.2f})")
plt.xlabel('Termination Count (per workweek)')
plt.ylabel('Average DPH (per workweek)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
